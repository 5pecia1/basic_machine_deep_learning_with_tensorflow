# Hypothesis and cost fucntion

$$H(x) = Wx + b$$
$$cost(W,b) = \frac{1}{m}\sum\limits_{i=1}^m (H(x^{(i)}) - y^{(i)})^2$$

* 학습에서의 목표  
  cost를 가장 작게하는 W, b를 구하는것이다

```python
import tensorflow as tf
```

학습 데이터를 준다

```python
x_data = [1,2,3,4]
y_data = [1,2,3,4]
```

$W$와 $b$의 값을 variable로 지정한다  
variable로 지정을 해야 나중에 모델이 업데이트를 할 수 있다.  
텐서플로우가 자동적으로 찾아올 수 있는지를 테스트 해보기 위해서 처음에는 $W$와 $b$에 랜덤한 값(-1 ~ 1)을 준다

```python
# Try to find values for W and b that compute y_data = w * x_data + b
# (We know that W should be 1 and 0, but Tensorflow will
# figure that out for us.)
W = tf.Variable(tf.random_uniform([1], -1.0, 1.0))
b = tf.Variable(tf.random_uniform([1], -1.0, 1.0))
```

$$H(x) = Wx + b$$

```python
# Our hypothesis
hypothesis = W * x_data + b
```

$$cost(W,b) = \frac{1}{m}\sum\limits_{i=1}^m (H(x^{(i)}) - y^{(i)})^2$$
`.reduce_mean()`을 사용한다고 계산이 일어나는 것은 아니다.

```python
# Simplified cost function
cost = tf.reduce_mean(tf.square(hypothesis - y_data))
```

여기서는 `.GradientDescentOpimizer()`를 이용해서 cost를 최소환 한다고 생각하면 된다.

```python
# Minimize
a = tf.Variable(0.1) # Learning rate, alpha
optimizer = tf.train.GradientDescentOptimizer(a)
train = optimizer.minimize(cost)
```

변수를 초기화 시킨다  
초기화 시키지 않으면 에러가 난다.

```python
# Simplified cost functino
init = tf.global_variables_initializer()
```

```python
# Launch the graph
sess = tf.Session()
```

첫번째로 실행을 시켜주어야 한다.

```python
sess.run(init)
```

loop를 2000번 돌면서
w, b를 찾는다
20번 마다 cost,w,b 값이 어떻게 되는지 보여준다

```python
for step in range(2001):
    sess.run(train)
    if step % 20 == 0:
        print (step, sess.run(cost), sess.run(W), sess.run(b))
```

### Placeholder를 이용해 보자

```python
X = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)
```

```python
# Our hypothesis
hypothesis = W * X + b

# Simplified cost function
cost = tf.reduce_mean(tf.square(hypothesis - Y))

# Minimize
a = tf.Variable(0.1) # Learning rate, alpha
optimizer = tf.train.GradientDescentOptimizer(a)
train = optimizer.minimize(cost)

# Simplified cost functino
init = tf.global_variables_initializer()

# Launch the graph
sess = tf.Session()
sess.run(init)

for step in range(2001):
    sess.run(train, feed_dict={X:x_data, Y:y_data})
    if step % 20 == 0:
        print (step, sess.run(cost, feed_dict={X:x_data, Y:y_data}), sess.run(W), sess.run(b))
```

예측하는 모델로 사용한다.  
모델을 작성 할 필요가 없다

X 값이 얼마일때 Y값은 얼마일까?

```python
# learns best fit is W:[1], b= [0]
print (sess.run(hypothesis, feed_dict={X:5}))
print (sess.run(hypothesis, feed_dict={X:2.5}))
```
